# Procedure:

Problem: Solve the Tower of Hanoi where there will be 3 three towers; A,B and C respectively and three blocks placed in A. The goal is to transfer the three blocks from A to C by the following constraits:

* Only one disk can be trasfer at a time
* Each move consists of taking the upper disk from one of the peg and placing it on the top of another peg
* Never a larger disk is placed on a smaller disk during the transfer.

To solve this problem:
- Move disc A to tower C
- Move disc B to tower B
- Move disc A to tower B
- Move disc C to tower C
- Move disc A to tower A
- Move disc B to tower C
- Move disc A to tower C

To solve this in computational thinking with python, we create an recursive function that has 4 parameters:
* n for number of disks
* Sour = "A"
* Aux = "B"
* Des = "C"


```python
def TOH(n, Sour, Aux, Des): # define the function and add 4 parameters
  if n==1:
    print("Move Disk", n ," from ", Sour ," to ",Des) # print
  else:
    TOH(n-1, Sour, Des, Aux)
    print("Move Disk", n ," from ", Sour ," to ",Des)
    TOH(n-1,  Aux, Sour, Des)

n = int(input("Enter n: ")) # input n
TOH(n, "A", "B", "C") # run the function
```
